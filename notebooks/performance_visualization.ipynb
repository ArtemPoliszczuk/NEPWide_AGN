## Metric plot for 1st iteration. No instance-weighting:

**2. Load data and create dataframes of specific form:**

```python
def get_df_for_plot(df, metrics, classifier):
    
    df_plot = df.T
    df_plot.rename(columns={0: "value"}, inplace=True)
    df_plot["classifier"] = classifier
    df_plot = df_plot.loc[metrics].copy()
    df_plot.reset_index(inplace=True)
    df_plot.rename(columns={"index": "metric"}, inplace=True)
    
    return df_plot
```

```python
def change_metric_names(df):
    df.loc[(df.metric == 'balanced_accuracy'),'metric']='bACC'
    df.loc[(df.metric == 'f1'),'metric']='F1'
    df.loc[(df.metric == 'pr_auc'),'metric']='PR AUC'
    df.loc[(df.metric == 'matthews_corrcoef'),'metric']='mCOR'
    df.loc[(df.metric == 'precision'),'metric']='Precision'
    df.loc[(df.metric == 'recall'),'metric']='Recall'
```

```python
def df_transformation(path, clf_name, class_balance, fuzzy, metrics_to_choose):
    
    if class_balance is None:
        clf_path_name = clf_name+ "/" + fuzzy
        legend_name = clf_name
        
    elif class_balance=="balanced":
        clf_path_name = clf_name + "_" + class_balance + "/" + fuzzy
        legend_name = "balanced " + clf_name
    
    elif class_balance=="not_balanced":
        clf_path_name = clf_name + "_" + class_balance + "/" + fuzzy
        legend_name = clf_name        
    
    result_path = os.path.join(path, clf_path_name)
    metrics = pd.read_csv(os.path.join(result_path, fuzzy+"_metrics.csv"), index_col=[0])
    plot_df = get_df_for_plot(metrics, metrics_to_choose, legend_name)
    change_metric_names(plot_df)
    
    return plot_df
```

```python
logistic_not_balanced = df_transformation(all_results_path, "logistic", "not_balanced", "normal", metrics_to_choose)
logistic_balanced = df_transformation(all_results_path, "logistic", "balanced", "normal", metrics_to_choose)
logistic_df = pd.concat([logistic_not_balanced, logistic_balanced])
```

```python
svm_not_balanced = df_transformation(all_results_path, "svm", "not_balanced", "normal", metrics_to_choose)
svm_balanced = df_transformation(all_results_path, "svm", "balanced", "normal", metrics_to_choose)
svm_df = pd.concat([svm_not_balanced, svm_balanced])
```

```python
rf_not_balanced = df_transformation(all_results_path, "rf", "not_balanced", "normal", metrics_to_choose)
rf_balanced = df_transformation(all_results_path, "rf", "balanced", "normal", metrics_to_choose)
rf_df = pd.concat([rf_not_balanced, rf_balanced])
```

```python
et_not_balanced = df_transformation(all_results_path, "et", "not_balanced", "normal", metrics_to_choose)
et_balanced = df_transformation(all_results_path, "et", "balanced", "normal", metrics_to_choose)
et_df = pd.concat([et_not_balanced, et_balanced])
```

```python
xgb_not_balanced = df_transformation(all_results_path, "xgb", "not_balanced", "normal", metrics_to_choose)
xgb_balanced = df_transformation(all_results_path, "xgb", "balanced", "normal", metrics_to_choose)
xgb_df = pd.concat([xgb_not_balanced, xgb_balanced])
```

```python
dummy_df = df_transformation(all_results_path, "dummy", None, "normal", metrics_to_choose)
hard_voter_df = df_transformation(all_results_path, "hard_voting_classifier", None, "normal", metrics_to_choose)
stacked_classifier_df = df_transformation(all_results_path, "stacked_classifier", None, "normal", metrics_to_choose)
```

# Instance-weighting plots. 1st iteration:

```python
def fuzzy_df_transformation(path, clf_name, class_balance, fuzzy):
    
    if class_balance is None:
        clf_path_name = clf_name+ "/" + fuzzy
        legend_name = clf_name
        
    elif class_balance=="balanced":
        clf_path_name = clf_name + "_" + class_balance + "/" + fuzzy
        legend_name = "balanced " + clf_name
    
    elif class_balance=="not_balanced":
        clf_path_name = clf_name + "_" + class_balance + "/" + fuzzy
        legend_name = clf_name        
    
    result_path = os.path.join(path, clf_path_name)
    metrics = pd.read_csv(os.path.join(result_path, fuzzy+"_metrics.csv"), index_col=[0])
    
    metrics["classifier"] = legend_name
    
    if fuzzy=="normal":
        metrics["type"] = "normal"
    elif fuzzy=="fuzzy_dist":
        metrics["type"] = "distance"
    elif fuzzy=="fuzzy_err":
        metrics["type"] = "error"
        
    return metrics
```

```python
svm_balanced_normal = fuzzy_df_transformation(all_results_path, "svm", "balanced", "normal")
svm_balanced_dist = fuzzy_df_transformation(all_results_path, "svm", "balanced", "fuzzy_dist")
svm_balanced_err = fuzzy_df_transformation(all_results_path, "svm", "balanced", "fuzzy_err")

fuzzy_svm = pd.concat([svm_balanced_normal, svm_balanced_dist, svm_balanced_err])
```

```python
rf_not_balanced_normal = fuzzy_df_transformation(all_results_path, "rf", "not_balanced", "normal")
rf_not_balanced_dist = fuzzy_df_transformation(all_results_path, "rf", "not_balanced", "fuzzy_dist")
rf_not_balanced_err = fuzzy_df_transformation(all_results_path, "rf", "not_balanced", "fuzzy_err")

rf_balanced_normal = fuzzy_df_transformation(all_results_path, "rf", "balanced", "normal")
rf_balanced_dist = fuzzy_df_transformation(all_results_path, "rf", "balanced", "fuzzy_dist")
rf_balanced_err = fuzzy_df_transformation(all_results_path, "rf", "balanced", "fuzzy_err")

fuzzy_rf = pd.concat([rf_not_balanced_normal, rf_not_balanced_dist, rf_not_balanced_err,
                      rf_balanced_normal, rf_balanced_dist, rf_balanced_err])
```

```python
et_not_balanced_normal = fuzzy_df_transformation(all_results_path, "et", "not_balanced", "normal")
et_not_balanced_dist = fuzzy_df_transformation(all_results_path, "et", "not_balanced", "fuzzy_dist")
et_not_balanced_err = fuzzy_df_transformation(all_results_path, "et", "not_balanced", "fuzzy_err")

et_balanced_normal = fuzzy_df_transformation(all_results_path, "et", "balanced", "normal")
et_balanced_dist = fuzzy_df_transformation(all_results_path, "et", "balanced", "fuzzy_dist")
et_balanced_err = fuzzy_df_transformation(all_results_path, "et", "balanced", "fuzzy_err")

fuzzy_et = pd.concat([et_not_balanced_normal, et_not_balanced_dist, et_not_balanced_err,
                      et_balanced_normal, et_balanced_dist, et_balanced_err])
```

```python
xgb_not_balanced_normal = fuzzy_df_transformation(all_results_path, "xgb", "not_balanced", "normal")
xgb_not_balanced_dist = fuzzy_df_transformation(all_results_path, "xgb", "not_balanced", "fuzzy_dist")
xgb_not_balanced_err = fuzzy_df_transformation(all_results_path, "xgb", "not_balanced", "fuzzy_err")

xgb_balanced_normal = fuzzy_df_transformation(all_results_path, "xgb", "balanced", "normal")
xgb_balanced_dist = fuzzy_df_transformation(all_results_path, "xgb", "balanced", "fuzzy_dist")
xgb_balanced_err = fuzzy_df_transformation(all_results_path, "xgb", "balanced", "fuzzy_err")

fuzzy_xgb = pd.concat([xgb_not_balanced_normal, xgb_not_balanced_dist, xgb_not_balanced_err,
                      xgb_balanced_normal, xgb_balanced_dist, xgb_balanced_err])
```

```python
all_fuzzy_df = pd.concat([fuzzy_svm, fuzzy_rf, fuzzy_et, fuzzy_xgb]) 
```
